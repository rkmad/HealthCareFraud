## Levene test for equal variances 'InscClaimAmtReimbursed_sum'

```python
# The Levene test tests the null hypothesis that all input samples are from populations with equal variances.
# ‘median’ : Recommended for skewed (non-normal) distributions>
scipy.stats.levene(trIeng_df[trIeng_df['PotentialFraud']==0]['InscClaimAmtReimbursed_sum'], 
                   trIeng_df[trIeng_df['PotentialFraud']==1]['InscClaimAmtReimbursed_sum'], center='median')
```

```python
# The small p-value suggests that the populations do not have equal variances.
# This is not surprising, given the sample variances:
```

```python
np.var(trIeng_df[trIeng_df['PotentialFraud']==0]['InscClaimAmtReimbursed_sum'], ddof=1)
```

```python
np.var(trIeng_df[trIeng_df['PotentialFraud']==1]['InscClaimAmtReimbursed_sum'], ddof=1)
```
